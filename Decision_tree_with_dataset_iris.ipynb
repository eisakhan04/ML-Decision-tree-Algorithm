```python
# import library
import numpy as np
import pandas as pd 
```

```python
# load dataset
from sklearn.datasets import load_iris
data = load_iris()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
df = pd.DataFrame(np.c_[data.data,data.target] , columns=[list(data.feature_names )+['target']]) 
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
x = df.iloc[ : , 0:-1 ]
y = df.iloc[ : , -1 ]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test =  train_test_split(x , y , test_size=0.2 , random_state=42)
 
print("shape of X_train :" , X_train.shape)
print("shape of X_test :" , X_test.shape)
print("shape of y_train :" , y_train.shape)
print("shape of y_test :" , y_test.shape)
```

# Train Decision Tree Classification Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier(criterion='gini')
DTC.fit(X_train , y_train)
```

```python
DTC.score(X_test , y_test)
```

```python
y_prd = DTC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test ,y_prd)
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80 , 50

plot_tree(DTC)
```

```python
DTC_entropy = DecisionTreeClassifier(criterion='entropy')
DTC_entropy.fit(X_train , y_train)
```

```python
DTC_entropy.score(X_test , y_test)
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80 , 50

plot_tree(DTC_entropy)
```

# Feature Scale

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train_sc = sc.transform(X_train)

X_test_sc = sc.transform(X_test)
```

```python
DTC_sc = DecisionTreeClassifier(criterion='gini')

DTC_sc.fit(X_train_sc , y_train)

DTC_sc.score(X_test_sc , y_test)
```

```python
DTC_sc = DecisionTreeClassifier()

DTC_sc.fit(X_train_sc , y_train)

DTC_sc.score(X_test_sc , y_test)
```

```python
DTC_sc.predict(X_test_sc)
```
